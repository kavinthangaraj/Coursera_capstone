### Getting the data

```python
# importing the required libraries

import pandas as pd
import numpy as np
```

```python
# parsing table data from wikipedia using pandas 
# creating a dataframe 'df' with the obtained table

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',header=0)[0]
print(df.info())
df.head()
```

### Cleaning the data

```python
# finding the non-assigned Boroughs

print(df['Borough'].value_counts())
```

```python
df = df[df['Borough'] != 'Not assigned'].reset_index(drop=True)  # removing the entries with non-assigned Boroughs
df['Borough'].value_counts()
```

```python
# finding the non-assigned Neighborhoods
    
df[df['Neighbourhood'] == 'Not assigned']    # only one non-assigned entry
```

```python
# replacing the non-assigned neighborhood with corresponding borough value
# using 'where' function from numpy  # method1

df['Neighbourhood'] = np.where(df['Neighbourhood'] == 'Not assigned', df['Borough'], df['Neighbourhood'])
df.head(9)

# method 2

# df['Neighbourhood'].iloc[6] = df['Borough'].iloc[6]
# df[df['Neighbourhood'] == 'Not assigned']
```

```python
# grouping values based on postcodes

df = df.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(', '.join).reset_index()
```

All the information that we need are under the 'items' key in the acquired 'json'. Let's define a function to extract categories and location of the venues

```python
# a function to extract the required information in this case, "name, category, lat and long data"

def get_category(entry):
    try:
        categories_list = entry['categories']
    except:
        categories_list = entry['venue.categories']
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

## Exploring all the neighbourhood around the main Toronto area

Now we explore the entire Toronto neighbourhood (from the toronto neighbourhood)

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            limit)
            
        # make the GET request
        
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Postcode', 
                  'Neighbourhood Latitude', 
                  'Neighbourhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
# calling the function to create the new dataframe 

toronto_venues = getNearbyVenues(names=toronto['Postcode'],
                                   latitudes=toronto['Latitude'],
                                   longitudes=toronto['Longitude'])
```

## Clustering the neighbourhood

Before clustering, we analyze the neighbourhood data to check for the most common type of venues in each neighbourhood

```python
# checking the number of unique categories

print(f"There are {len(toronto_venues['Venue Category'].unique())} unique categories")
```

We use one hot encoding, which is one of the methods to convert categorical variables to numerical values (0s and 1s)

```python
# by applying one hot encoding on categories using pandas.get_dummies method, we analyze each neighbourhood

toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# adding neighborhood column to the one hot encoded dataframe

toronto_onehot['Postcode'] = toronto_venues['Postcode'] 

# moving the neighborhood column to the first column

fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

print(toronto_onehot.shape)
toronto_onehot.head()
```

Now, we group the neighbourhood and analyze the frequency (popularity) of each category in that particular neighbourhood

```python
toronto_grouped = toronto_onehot.groupby(by=['Postcode']).mean().reset_index()
toronto_grouped
```

```python
toronto_grouped.shape
```

#### Creating a dataframe by most common venues in each neighbourhood

```python
# define a function to sort the venues in descending order based on frequency of occurence

def most_common_venues(entry, top_venues):
    entry_categories = entry.iloc[1:]
    entry_categories_sorted = entry_categories.sort_values(ascending=False)
    
    return entry_categories_sorted.index.values[0:top_venues]
```

```python
num = 10   # top 10 venues

indicators = ['st', 'nd', 'rd']  # naming the columns

# create columns according to number of top venues

columns = ['Postcode']
for ind in np.arange(num):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe

neighbourhood_venues_sorted = pd.DataFrame(columns=columns)
neighbourhood_venues_sorted['Postcode'] = toronto_grouped['Postcode']

#calling the function

for ind in np.arange(toronto_grouped.shape[0]):
    neighbourhood_venues_sorted.iloc[ind, 1:] = most_common_venues(toronto_grouped.iloc[ind, :], num)

neighbourhood_venues_sorted.head()
```

### Clustering the neighbourhood 

We use the common clustering algorithms *k-means clustering* to cluster the neighbourhood

```python
# loading the necessary libraries
from sklearn.cluster import KMeans
```

```python
# create an instance for KMeans

num_clusters = 3    # number of clusters

toronto_grouped_clustering = toronto_grouped.drop(['Postcode'], axis=1)

kcluster = KMeans(n_clusters=num_clusters, random_state=7).fit(toronto_grouped_clustering)
```

```python
kcluster.labels_
```

```python
# merging the toronto_dataframe with cluster labels and most common categories

neighbourhood_venues_sorted.insert(0, 'Cluster Labels', kcluster.labels_)

toronto_merged = toronto

toronto_merged = toronto_merged.join(neighbourhood_venues_sorted.set_index('Postcode'), on='Postcode')

toronto_merged.head()
```

### Visualize the clusters

```python
# import the libraries for plotting

import matplotlib.cm as cm
import matplotlib.colors as colors

# create a map of toronto

map_clusters = folium.Map(location=[lat, lon], zoom_start=12)

# color scheme for clusters

x = np.arange(num_clusters)
ys = [i + x + (i*x)**2 for i in range(num_clusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map

markers_colors = []
for lat, lng, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Postcode'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.9).add_to(map_clusters)
       
map_clusters
```
